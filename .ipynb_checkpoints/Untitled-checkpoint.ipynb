```python
import tensorflow as tf
```

```python
# Lookup table 만드는 부분

seq_len = 10 # 입력된 문장의 길이
k = 3

rangeVec = tf.range(seq_len)
rangeMat = tf.reshape(tf.tile(rangeVec, [seq_len]), [seq_len, seq_len])
distance_mat = rangeMat - tf.transpose(rangeMat)
distance_mat_clipped = tf.clip_by_value(distance_mat, -k, k)
final_mat = distance_mat_clipped + k
final_mat
```

```python
depth = 2 # key의 차원 수
vocab_size = k * 2 + 1

embeddings_table = tf.compat.v1.get_variable("embeddings", [vocab_size, depth])
embeddings = tf.gather(embeddings_table, final_mat)
embeddings_table
```

```python
embeddings[0]
```

```python
qk = tf.compat.v1.get_variable("embeddings", [3, 8, seq_len, depth])
x_t = tf.transpose(qk, [2, 0, 1, 3])
x_t_r = tf.reshape(x_t, [seq_len, 8 * 3, -1])
x_tz_matmul = tf.matmul(x_t_r, embeddings, transpose_b=True)
x_tz_matmul_r = tf.reshape(x_tz_matmul, [seq_len, 3, 8, -1])
x_tz_matmul_r_t = tf.transpose(x_tz_matmul_r, [1, 2, 0, 3])
x_tz_matmul_r_t
```
